```python
from collections import Counter

def get_top_n_words(words, n):
    counter = Counter(words)
    return counter.most_common(n)

# Example usage
n = 100  # Change this value to get a different number of top words
top_n_words_pence = get_top_n_words(words_pence, n)
top_n_words_harris = get_top_n_words(words_harris, n)
print(f'Top words from pence {top_n_words_pence}')
print(f'Top words from harris {top_n_words_harris}')
```
